```python
import random
import json

first_names = [
    "mahesh", "kamal", "nimal", "suresh", "chamidu", "ruwan", "naveen", "pathum", "sisitha", "charuka",
    "tharindu", "dilshan", "supun", "isuru", "ranil", "janith", "kasun", "darshana", "lahiru", "vimukthi",
    "isuru", "vinnath", "tharidu", "dulan", "samidu", "shan"
]

middle_names = [
    "kumara", "chathuranga", "lakshan", "pradeep", "indika", "nuwan", "vinnath", "nilan",
    "sandun", "madushan", "sachintha", "harsha", "buddhika", "chamara", "malith", "shehan",
    "nadeeshan", "siranga", "ninura", "madawa", "sadaruwan", "sampath"
]

last_names = [
    "perera", "silva", "fernando", "wijesinghe", "bandara", "liyanage", "madanayaka", "gamage",
    "jayasinghe", "de silva", "ranasinghe", "weerasinghe", "gunasekara", "abeywardana", "ratnayake",
    "satharasinghe", "marasingha", "samarakoon", "siriwardana"
]

addresses = [
    "23/1 kandy road colombo", "45a galle road matara", "12 temple street kandy",
    "67/2 lake view nuwaraeliya", "34 station road jaffna", "210/5 moraghahena road pannipitiya",
    "22/b rathmaldeniya road pannipitiya", "12 edmonton road colombo", "95/3 high level road maharagama",
    "30 park avenue dehiwala", "77/1 station road negombo", "11/2 old dutch street galle",
    "56 temple road anuradhapura", "81 a9 road trincomalee", "102/7 lake view road badulla",
    "37 airport road ratmalana", "205 main street kurunegala", "40/8 gampaha road minuwangoda",
    "10/5 colombo road kalutara", "88/3 new town road hambantota", "63 hill street pannala",
    "14/9 kaduwela road battaramulla", "278 rathmaldeniya road pannipitiya"
]


def generate_license_data(num_records=10):
    data = []
    for _ in range(num_records):
        text = "driving licence democratic socialist republic of sri lanka "
        entities = []

        id_number = f"{random.randint(100000000, 999999999)}" + ("v" if random.choice([True, False]) else "")
        license_number = f"{random.randint(10000000, 99999999)}"
        name_parts = [random.choice(first_names)] + random.sample(middle_names, random.randint(0, 2)) + [random.choice(last_names)]
        name = " ".join(name_parts)
        address = random.choice(addresses)
        dob = f"{random.randint(1, 28):02d}.{random.randint(1, 12):02d}.{random.randint(1970, 2005)}"
        issue_date = f"{random.randint(1, 28):02d}.{random.randint(1, 12):02d}.{random.randint(2010, 2022)}"
        expiry_date = f"{random.randint(1, 28):02d}.{random.randint(1, 12):02d}.{random.randint(2030, 2045)}"
        blood_group = random.choice(["a+", "a-", "b+", "b-", "o+", "o-", "ab+", "ab-"])

        fields = [
            ("ID", id_number),
            ("LICENSE_NUMBER", license_number),
            ("NAME", name),
            ("ADDRESS", address),
            ("DOB", dob),
            ("ISSUE_DATE", issue_date),
            ("EXPIRY_DATE", expiry_date)
        ]

        for label, value in fields:
            if random.random() < 0.5 or not value.strip():
                continue
            start = len(text)
            text += value + " "
            end = len(text) - 1
            entities.append({"start": start, "end": end, "label": label})

        if random.random() < 0.5 or not blood_group.strip():
            text += "blood group "
        else:
            text += "blood group "
            start = len(text)
            text += blood_group + " "
            end = len(text) - 1
            entities.append({"start": start, "end": end, "label": "BLOOD_GROUP"})

        text += "j.m.u.k jayasekera commissioner general motor traffic"

        data.append({
            "text": text.strip(),
            "entities": entities
        })

    return data


dataset = generate_license_data(2000)

with open("license_data_with_missing_parts.json", "w") as f:
    json.dump(dataset, f, indent=4)
```
